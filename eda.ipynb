```python
import pandas as pd

# Read the CSV files with specified encoding
df1 = pd.read_csv("FastFoodNutritionMenuV2.csv", encoding='latin1')
df2 = pd.read_csv("food data hand made(1).csv", encoding='latin1')
```

```python

columns_to_convert = df1.columns[1:]

# Convert selected columns to float with error handling
for col in columns_to_convert:
    df1[col] = pd.to_numeric(df1[col], errors='coerce')

print("Data types of each column after conversion:")
print(df1.dtypes)
print(df2.dtypes)
```
