```python
import pandas as pd
import matplotlib.pyplot as plt
A=pd.read_csv('cleaned_data_8k.csv')
A=A.dropna()
A
```

```python
thresholds = {
    'Energy(kcal)_per 100 g': 400,
    'Protein(g) per 100 g': 10,
    'Carbohydrate(g) per 100 g': 50,
    'total sugar(g) per 100 g': 20,
    'Added  Sugar(g) per 100 g': 8,
    'trans fat (g) _per 100 g': 5,
    'total fat (g) per 100 gms': 30,
    'Sodium(mg) per 100 g': 400,
    'iron (mg) per 100 gm': 2,
    'Cholesterol per 100 mg': 20
}

def label_health(row):
    total_thresholds = len(thresholds)
    crossed_count = sum(row[col] > threshold for col, threshold in thresholds.items())
    
    # Calculate the percentage of thresholds crossed
    percentage_crossed = (crossed_count / total_thresholds) * 100
    
    # Assign labels based on the percentage crossed
    if percentage_crossed == 0:
        return '100% Healthy'
    elif percentage_crossed == 100:
        return '100% Unhealthy'
    elif percentage_crossed < 50:
        return ' 50% Unhealthy'
    else:
        return ' 50% Unhealthy'

# Apply the function to each row in the DataFrame to label the foods
A['Health_Label'] = A.apply(label_health, axis=1)

# Display the updated DataFrame
A
```

```python
Health_counts = A['Health_Label'].value_counts()
Health_counts
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Drop unnecessary columns
X = A.drop(columns=['Food_name', 'Cluster', 'Health_Label'])
y = A['Health_Label']

# Convert categorical variables to numerical representations if needed

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize models
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC()
}

# Train and evaluate models
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy

# Print accuracy of each model
for name, accuracy in results.items():
    print(f'{name}: {accuracy}')
```
